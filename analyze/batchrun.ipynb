```python
from traffic_model.simulation.model import *
from mesa.batchrunner import BatchRunner
```

```python
max_steps = 72000
start_steps = [252000, 468000, 576000]
dfs = []
index = 0
make_df = {
    'light_12': [],
    'start_steps': [],
    'avg_car_steps': [],
    'finished_car_wait': []
}
for start_step in start_steps:
    for light_timing in range(0, 21, 10):
        model = Traffic(light_12=light_timing,
                        width=750,
                        height=750,
                        max_steps=max_steps,
                        start=start_step)
        for i in range(max_steps):
            model.step()
        index+=1
        print(index)
        make_df['start_steps'].append(start_step)
        make_df['avg_car_steps'].append(model.finished_car_steps_int)
        make_df['finished_car_wait'].append(model.finished_car_wait_int)
        make_df['light_12'].append(light_timing)
```
